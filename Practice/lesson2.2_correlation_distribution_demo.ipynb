
# Lesson 2: Selecting a Chart Type for Correlation & Distribution

In this lesson, we focus on selecting appropriate chart types when analyzing **distributions** and **relationships/correlations** in data.

We will cover:
- **Distribution charts**: Histogram, Density plot, Boxplot, Violin, Distribution comparisons
- **Correlation/relationship charts**: Scatterplot, Scatterplot matrix, Bubble chart, Gaussian density scatterplot, Parallel coordinates, Radar chart, Bullet chart, Heatmap (correlation matrix)
- **Geospatial visualization**: Choropleth filled-map

Each example uses **real-world datasets** or synthetic distributions. 


```

# Imports
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.preprocessing import MinMaxScaler

# Datasets from seaborn
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
penguins = sns.load_dataset("penguins")

# For reproducibility
np.random.seed(42)
```

## Histogram

```

plt.figure(figsize=(6,4))
sns.histplot(tips['total_bill'], bins=20, kde=False, color="skyblue")
plt.title("Histogram of Total Bill")
plt.show()
```

## Density Plot (KDE)

```

plt.figure(figsize=(6,4))
sns.kdeplot(tips['total_bill'], fill=True, color="orange")
plt.title("Density Plot of Total Bill")
plt.show()
```

## Boxplot (with outliers)

```

plt.figure(figsize=(6,4))
sns.boxplot(x="day", y="total_bill", data=tips, palette="Set2")
plt.title("Boxplot of Total Bill by Day")
plt.show()
```

## Violin Plot

```

plt.figure(figsize=(6,4))
sns.violinplot(x="day", y="total_bill", data=tips, palette="Set3")
plt.title("Violin Plot of Total Bill by Day")
plt.show()
```

## Comparing Distributions: Normal, Skewed, Bimodal

```

data_normal = np.random.normal(loc=0, scale=1, size=1000)
data_skewed = np.random.exponential(scale=1, size=1000)
data_bimodal = np.concatenate([np.random.normal(-2, 0.5, 500), np.random.normal(3, 0.5, 500)])

plt.figure(figsize=(8,6))
sns.kdeplot(data_normal, fill=True, label="Normal")
sns.kdeplot(data_skewed, fill=True, label="Skewed")
sns.kdeplot(data_bimodal, fill=True, label="Bimodal")
plt.title("Different Types of Distributions")
plt.legend()
plt.show()
```

## Scatterplot

```

plt.figure(figsize=(6,4))
sns.scatterplot(x="total_bill", y="tip", hue="smoker", data=tips, palette="Set1")
plt.title("Scatterplot: Total Bill vs Tip")
plt.show()
```

## Scatterplot Matrix (Pairplot)

```

sns.pairplot(iris, hue="species", diag_kind="kde")
plt.show()
```

## Bubble Chart

```

plt.figure(figsize=(6,4))
plt.scatter(tips["total_bill"], tips["tip"], s=tips["size"]*50, alpha=0.5)
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Bubble Chart: Bill vs Tip (size = party size)")
plt.show()
```

## Gaussian Density Scatterplot (2D KDE)

```

plt.figure(figsize=(6,4))
sns.kdeplot(x=tips["total_bill"], y=tips["tip"], fill=True, cmap="Blues")
sns.scatterplot(x="total_bill", y="tip", data=tips, alpha=0.4)
plt.title("2D Density Scatterplot")
plt.show()
```

## Parallel Coordinates

```

from pandas.plotting import parallel_coordinates

iris_scaled = iris.copy()
iris_scaled[iris_scaled.columns[:-1]] = MinMaxScaler().fit_transform(iris_scaled[iris_scaled.columns[:-1]])

plt.figure(figsize=(8,6))
parallel_coordinates(iris_scaled, "species", colormap=plt.cm.Set1)
plt.title("Parallel Coordinates - Iris Dataset")
plt.show()
```

## Radar Chart

```

categories = list(iris.columns[:-1])
N = len(categories)
values = iris.iloc[0, :-1].values.flatten().tolist()
values += values[:1]  # close loop

angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

plt.figure(figsize=(6,6))
ax = plt.subplot(111, polar=True)
plt.xticks(angles[:-1], categories)
ax.plot(angles, values, linewidth=2, linestyle='solid')
ax.fill(angles, values, 'skyblue', alpha=0.4)
plt.title("Radar Chart Example (First Iris Sample)")
plt.show()
```

## Bullet Chart

```

# Simple bullet chart implementation
actual = 70
target = 90

plt.figure(figsize=(6,2))
plt.barh(["Metric"], actual, color="skyblue")
plt.axvline(target, color="red", linestyle="--", label="Target")
plt.title("Bullet Chart Example")
plt.legend()
plt.show()
```

## Heatmap of Correlation Matrix

```

corr = tips.corr(numeric_only=True)
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap - Tips Dataset")
plt.show()
```

## Choropleth Map Example (GDP per Capita)

```

# Example dataset from plotly
gapminder = px.data.gapminder().query("year==2007")

fig = px.choropleth(gapminder, locations="iso_alpha", color="gdpPercap",
                    hover_name="country", color_continuous_scale=px.colors.sequential.Plasma)
fig.show()
```
