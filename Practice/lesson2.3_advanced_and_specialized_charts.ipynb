## 1. Setup and Imports

```

import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import yfinance as yf

pd.set_option('display.max_columns', 50)
print("Libraries imported successfully!")
```


## 2. Choropleth Map – COVID-19 Data

We will visualize **COVID-19 total cases per million** across countries.  
Data comes directly from [Our World in Data](https://ourworldindata.org/coronavirus).

Steps:
1. Load dataset directly from OWID.
2. Filter to latest available date.
3. Create a world map with color intensity by cases per million.


```

url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
covid = pd.read_csv(url)
covid['date'] = pd.to_datetime(covid['date'])

# Keep only latest data for each country
latest = covid.groupby('iso_code').apply(lambda x: x.loc[x['date'].idxmax()])
latest = latest[latest['iso_code'].str.len() == 3]

fig = px.choropleth(
    latest,
    locations='iso_code',
    color='total_cases_per_million',
    hover_name='location',
    color_continuous_scale='Reds',
    title='COVID-19 Total Cases per Million (Latest Available)'
)
fig.show()
```


## 3. Stock Charts – Financial Data Visualization

Let’s visualize **Apple (AAPL)** stock performance using **candlestick** and **moving average** charts.

Steps:
1. Download daily stock data for 1 year with `yfinance`.
2. Create **Candlestick** and **OHLC** plots.
3. Add **Moving Averages (MA50, MA200)** for trend insight.


```

data = yf.download("AAPL", period="1y", interval="1d")

fig = go.Figure(data=[go.Candlestick(
    x=data.index,
    open=data['Open'],
    high=data['High'],
    low=data['Low'],
    close=data['Close']
)])
fig.update_layout(title='AAPL Candlestick Chart', xaxis_title='Date', yaxis_title='Price (USD)')
fig.show()

# Moving averages
data['MA50'] = data['Close'].rolling(window=50).mean()
data['MA200'] = data['Close'].rolling(window=200).mean()

fig = go.Figure()
fig.add_trace(go.Scatter(x=data.index, y=data['Close'], mode='lines', name='Close'))
fig.add_trace(go.Scatter(x=data.index, y=data['MA50'], mode='lines', name='MA50'))
fig.add_trace(go.Scatter(x=data.index, y=data['MA200'], mode='lines', name='MA200'))
fig.update_layout(title='AAPL Stock Price with Moving Averages')
fig.show()
```


## 4. Waterfall Chart – Sequential Gains and Losses

A **Waterfall Chart** visualizes cumulative effects of sequential positive/negative values, e.g. company profit analysis.


```

x = ['Revenue', 'Costs', 'Operating Income', 'Taxes', 'Net Profit']
measure = ['relative', 'relative', 'relative', 'relative', 'total']
y = [1000, -400, -200, -150, 0]

fig = go.Figure(go.Waterfall(
    name="Company Performance",
    orientation="v",
    measure=measure,
    x=x,
    textposition="outside",
    y=y,
    connector={"line": {"color": "rgb(63, 63, 63)"}},
))
fig.update_layout(title="Waterfall Chart: Company Profit Breakdown")
fig.show()
```


## 5. Funnel and Sankey Diagrams – Flow Visualization

### 🧩 Funnel Chart
Shows how quantities decrease through sequential stages (e.g. marketing conversion).

### 🔀 Sankey Diagram
Shows flow between categories or stages (e.g. user paths, department transfers).


```

# Funnel
stages = ['Visitors', 'Leads', 'Marketing Qualified', 'Sales Qualified', 'Customers']
values = [10000, 6000, 3000, 1200, 400]

fig = go.Figure(go.Funnel(y=stages, x=values))
fig.update_layout(title='Marketing Funnel Conversion')
fig.show()

# Sankey
label = ["Website", "Email", "Ad", "Signup", "Purchase"]
source = [0, 0, 1, 2]
target = [3, 3, 4, 4]
value = [8, 4, 2, 6]

fig = go.Figure(go.Sankey(
    node=dict(label=label, pad=20, thickness=20),
    link=dict(source=source, target=target, value=value)
))
fig.update_layout(title_text="User Flow Sankey Diagram", font_size=12)
fig.show()
```


## 6. Treemap and Mosaic Charts – Hierarchical & Proportional Data

Treemaps display hierarchical categories as nested rectangles sized by value.  
We’ll use the **Titanic dataset** to illustrate proportions of passengers by class, sex, and survival.


```

titanic = px.data.titanic()

fig = px.treemap(titanic, path=['class', 'sex', 'survived'], title='Titanic Treemap: Class, Sex, and Survival')
fig.show()
```


## 7. Mixed / Combo Charts – Comparing Two Variables

We’ll combine a **bar chart** (new COVID-19 cases) and a **line chart** (new vaccinations) on a secondary y-axis.


```

covid_country = covid[(covid['location'] == 'United States') & (covid['date'] > '2022-01-01')]

fig = go.Figure()
fig.add_trace(go.Bar(x=covid_country['date'], y=covid_country['new_cases'], name='New Cases'))
fig.add_trace(go.Scatter(x=covid_country['date'], y=covid_country['new_vaccinations'], name='New Vaccinations', yaxis='y2'))

fig.update_layout(
    title='US COVID Cases vs Vaccinations (Mixed Chart)',
    yaxis=dict(title='New Cases'),
    yaxis2=dict(title='New Vaccinations', overlaying='y', side='right')
)
fig.show()
```
